```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

left_image = cv2.imread('/Users/only/Downloads/IMG_6659.jpg', cv2.IMREAD_GRAYSCALE)
right_image = cv2.imread('/Users/only/Downloads/IMG_6660.jpg', cv2.IMREAD_GRAYSCALE)
stereo = cv2.StereoBM_create(numDisparities=16*8, blockSize=15)
disparity = stereo.compute(left_image, right_image).astype(np.float32) / 16.0
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.title('Left Image')
plt.imshow(left_image, cmap='gray')
plt.subplot(1, 3, 2)
plt.title('Right Image')
plt.imshow(right_image, cmap='gray')
plt.subplot(1, 3, 3)
plt.title('Disparity Map')
plt.imshow(disparity, cmap='jet')

plt.show()

```
